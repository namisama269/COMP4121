```python
# The given hash function
def h(x, p):
    return x % p
```

```python
# Let's plot the hash function with p=31 and see the drawback visually
p = 31
x_values = [x for x in range(1, 101)]
y_values = [h(x, p) for x in range(1, 101)]
plt.scatter(x_values, y_values)
plt.show()
```

```python
# At every multiple of p=31, there is a large jump discontinuity.
# A possible way to fix this is to make the the first interval of 31
# increasing (/), then next interval decreasing (\), then we would get 
# a /\/\/\ shape instead of ///.

# To do this, we consider the modulus of 2p, if it is less than p then
# we keep the same function, but if it is larger then subtract it from
# 2p - 1.
# Note that we want h_ < p and not h_ <= p since at h_ = p, we cannot 
# return p since the range of possible outputs from the hash function
# are from 0 to p-1.
```

```python
def h2(x, p):
    h_ = x  % (2*p)
    if h_ < p:
        return h_
    else:
        return 2*p - 1 - h_
```

```python
x2 = [x for x in range(1, 101)]
y2 = [h2(x, p) for x in range(1, 101)]
plt.scatter(x2, y2)
print(y2)
plt.show()
```
