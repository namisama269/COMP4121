```python
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
def factorial(n):
    if n <= 1:
        return 1
    ans = 1
    for i in range(2, n+1):
        ans *= i
    return ans
```

```python
def ncr(n, r):
    return factorial(n) // (factorial(r) * factorial(n-r))
```

```python
# Plot for part (c)
```

```python
n = 100
x = np.array([i for i in range(n+1)])
y_vals = []
for k in range(n+1):
    val = 0
    for i in range(k, n+1):
        val += ncr(n, i)
    val *= 1/(2**n)
    y_vals.append(val)
y = np.array(y_vals)

sns.scatterplot(x, y)
```
