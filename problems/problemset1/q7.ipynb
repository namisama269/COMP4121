```python
import numpy as np
from scipy import linalg
from fractions import Fraction
```

```python
A = np.array([[4/5, 3/4], [1/5, 1/4]])
```

```python
# find eigenvalues and eigenvectors of A
evals, evecs = linalg.eig(A, left=False, right=True)
```

```python
print(evals)
print(evecs)
```

```python
# The first column is the eigenvector corresponding to eigenvalue 1
# normalise it so it sums to 1
ans = [evecs.T[0]]
ans /= np.sum(ans)
print(ans)
```

```python
c, t = ans[0]
```

```python
print(f"c: {str(Fraction(c).limit_denominator())}")
print(f"t: {str(Fraction(t).limit_denominator())}")
```
