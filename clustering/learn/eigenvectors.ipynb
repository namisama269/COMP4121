```python
import numpy as np
from scipy import linalg
```

```python
A = np.array([[1,2,3],[3,2,1],[1,0,-1]])
w, v = linalg.eig(A)
```

```python
print(w)
```

```python
print(v)
```

```python
print(v.T)
```

```python
lam = w[1]
u = v.T[1]
```

```python
print(np.dot(A,u))
```

```python
print(lam*u)
```
