```python
from dft import dft, idft, dft_convolution
from fft import fft, ifft, fft_convolution
from fft_normalized import fft as fft_norm, ifft as ifft_norm, \
fft_convolution as fft_conv_norm
from sequences import convolution
```

```python
def round_real_conv(conv):
    return [round(x.real, 2) for x in conv]
```

```python
a = [float(x) for x in input("Enter sequence a: ").split()]
b = [float(x) for x in input("Enter sequence b: ").split()]
actual = convolution(a, b)
print()

print("DFT convolution:")
print(round_real_conv(dft_convolution(a, b)))

print("FFT convolution:")
print(round_real_conv(fft_convolution(a, b)))

print("Normalized FFT convolution:")
print(round_real_conv(fft_conv_norm(a, b)))

print("Actual: ")
print(actual)
```
