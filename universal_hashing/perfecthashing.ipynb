```python
def find_table_size(n):
    """
    Find a prime number m such that n ≤ m ≤ 2n to be the size of the hash table
    """
    if n == 0 or n == 1:
        return 2
    for m in range(n, 2*n):
        if is_prime(m):
            return m
    return None
```

```python
n = 10
```

```python
m = find_table_size(n)
print(m)
```

```python
r = math.floor(math.log(n, m))
print(r)
```
